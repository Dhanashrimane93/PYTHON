```python
# Q.4 Are there any pattern that can be observed between total honey production and
# value of production every year?
```

```python
plt.figure(figsize=(25,10))
sns.barplot(x='year',y='prodvalue',data=df.sort_values('totalprod',ascending=False),
           label='production value',color='green',ci=None)
sns.barplot(x='year',y='totalprod',data=df.sort_values('totalprod',ascending=False),
           label='total production',color='red',ci=None,alpha=0.5)
plt.legend(ncol=2,loc='upper left',frameon=True)
plt.show()
```

```python
# Q.5 How has the value of production which in some sense could be tied to demand changed every year?
```

```python
plt.figure(figsize=(25,10))
sns.barplot(x='year',y='prodvalue',data=df.sort_values('prodvalue',ascending=False),
           color='red',ci=None,alpha=0.8)
plt.title('value of production based on year',size=25)
plt.show()
```
