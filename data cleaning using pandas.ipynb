```python
store_data={'date':['2022_02_10','2022_02_11','2022_02_13','2022_02_14',
                   '2022_02_15','2022_02_16'],
           'prod':['a','b','c','d','b','d'],
           'sales':['4.59','8.37','6.5','5.82','7.23','9.0']}
store_df = pd.DataFrame(store_data)
store_df
```

```python
store_df.dtypes
```

```python
store_df['date']=pd.to_datetime(store_df['date'])
store_df['sales']=pd.to_numeric(store_df['sales'])
```

```python
store_df.dtypes
```

```python
#changing index
store_df.head()
```

```python
store_df.set_index('date',inplace=True)
store_df.head()
```

```python
# renaming index

store_df.index.name='product'
store_df.head()
```
